```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from numpy.polynomial import polynomial as P
from sklearn.linear_model import LinearRegression #used to fit the linear regression model
from sklearn.model_selection import train_test_split  #used to split the data
```

```python
# import the file
df  = pd.read_csv('/content/drive/MyDrive/data/Real estate.csv')
df
```

```python
x = df.drop(['Y house price of unit area'], axis=1)
x
y = df[['Y house price of unit area']]
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y)
x_test
```

```python
x_train
```

```python
reg = LinearRegression()   #applying the function
reg.fit(x_train,y_train)
y_train_pred = reg.predict(x_train)
plt.scatter(x_train['X2 house age'], y_train)   #plotting the graph between one of the properties and y_train
plt.scatter(x_train['X2 house age'], y_train_pred)  #same with Y_train_pred
```

```python
y_test_pred = reg.predict(x_test)
plt.scatter(x_test['X2 house age'],y_test)
plt.scatter(x_test['X2 house age'],y_test_pred)
```

```python
from sklearn.metrics import mean_squared_error,r2_score  #calculating r2 score to check accuracy.
error = mean_squared_error(y_pred= y_train_pred, y_true = y_train)
score_train = r2_score(y_pred = y_train_pred,y_true = y_train)
score_test = r2_score(y_pred = y_test_pred,y_true = y_test)
print(score_test,score_train)
```
